```python
%Zad1 Stwórz bazę faktów dla drzewa genealogicznego, która obejmuje relacje rodzic-dziecko, małżeństwo oraz płeć. Następnie zdefiniuj reguły przodek(X, Y), brat(X, Y), siostra(X, Y), kuzyn(X, Y).
% Fakty
ojciec(jan, anna).
ojciec(jan, piotr).
ojciec(adam, pawel).
ojciec(adam, ewa).
matka(ela, pawel).
matka(ela, ewa).
matka(maria, anna).
matka(maria, piotr).
malzenstwo(jan, maria).
malzenstwo(adam, ela).
mezczyzna(jan).
mezczyzna(piotr).
mezczyzna(adam).
mezczyzna(pawel).
kobieta(maria).
kobieta(anna).
kobieta(ela).
kobieta(ewa).

% Reguły
przodek(X, Y) :- ojciec(X, Y).
przodek(X, Y) :- matka(X, Y).
przodek(X, Y) :- ojciec(X, Z), przodek(Z, Y).
przodek(X, Y) :- matka(X, Z), przodek(Z, Y).

brat(X, Y) :- mezczyzna(X), ojciec(Z, X), ojciec(Z, Y), X \= Y.
brat(X, Y) :- mezczyzna(X), matka(Z, X), matka(Z, Y), X \= Y.
brat(jan, adam).
brat(adam, jan).
brat(piotr, anna).
brat(pawel, ewa).

siostra(X, Y) :- kobieta(X), ojciec(Z, X), ojciec(Z, Y), X \= Y.
siostra(X, Y) :- kobieta(X), matka(Z, X), matka(Z, Y), X \= Y.
siostra(maria, ela).
siostra(ela, maria).
siostra(anna, piotr).
siostra(ewa, pawel).

% Kuzyn
kuzyn(X, Y) :-
    ojciec(Z, X), ojciec(W, Y), brat(Z, W), X \= Y.
kuzyn(X, Y) :-
    matka(Z, X), matka(W, Y), siostra(Z, W), X \= Y.
kuzyn(X, Y) :-
    ojciec(Z, X), matka(W, Y), brat(Z, W), X \= Y.
kuzyn(X, Y) :-
    matka(Z, X), ojciec(W, Y), siostra(Z, W), X \= Y.

% Lista kuzynów
kuzyni(X, Lista) :-
    findall(Y, kuzyn(Y, X), Lista).
```

```python
%Zad2
% Lotniska
lotnisko(warszawa).
lotnisko(krakow).
lotnisko(gdansk).
lotnisko(wroclaw).
lotnisko(poznan).

% Połączenia (z kosztami)
polaczenie(warszawa, krakow, 300).
polaczenie(krakow, gdansk, 200).
polaczenie(warszawa, gdansk, 500).
polaczenie(krakow, wroclaw, 150).
polaczenie(wroclaw, poznan, 250).
polaczenie(gdansk, poznan, 400).

% Sprawdzamy, czy istnieje bezpośrednie połączenie między dwoma lotniskami
polaczenie(X, Y) :- polaczenie(X, Y, _); polaczenie(Y, X, _).

% Określamy koszt podróży między lotniskami X i Y
koszt_podrozy(X, Y, C) :- polaczenie(X, Y, C).
koszt_podrozy(X, Y, C) :- polaczenie(Y, X, C).

% Najkrótsza droga między X i Y, szukamy rekurencyjnie połączeń i obliczamy najtańszą trasę
najkrotsza_droga(X, Y) :-
    najkrotsza_droga(X, Y, [X], 0, C),
    write('Najkrótsza droga kosztuje: '), write(C), nl.

najkrotsza_droga(X, X, _, C, C).
najkrotsza_droga(X, Y, Visited, Acc, C) :-
    polaczenie(X, Z, Cost),
    \+ member(Z, Visited),  % Sprawdzamy, czy już odwiedziliśmy to lotnisko
    NewAcc is Acc + Cost,
    najkrotsza_droga(Z, Y, [Z|Visited], NewAcc, C).
```

```python
Przykładowe wyniki:
?- polaczenie(warszawa, krakow).
Oczekiwany wynik: true
?- koszt_podrozy(warszawa, gdansk, C).
Oczekiwany wynik: C = 50
?- najkrotsza_droga(warszawa, gdansk).
Oczekiwany wynik: trasa przez Kraków, koszt 500
```

```python
%Zad3
% Deklaracja dynamicznych faktów
:- dynamic produkt/3.

% Fakty: Produkty w magazynie
produkt(tv, 10, sektor_a).
produkt(pralka, 5, sektor_b).
produkt(lodowka, 3, sektor_c).

% Reguła sprawdzająca dostępność produktu w magazynie
dostepny_produkt(X) :-
    produkt(X, Ilosc, _),
    Ilosc > 0.

% Reguła przenosząca produkt do innego sektora
przenies_produkt(X, Y) :-
    produkt(X, Ilosc, _),  % Sprawdza, czy produkt istnieje
    retract(produkt(X, Ilosc, _)),  % Usuwa stary fakt
    assertz(produkt(X, Ilosc, Y)),  % Dodaje nowy fakt z nowym sektorem
    write('produkt('), write(X), write(', '), write(Ilosc), write(', '), write(Y), write(')'), nl.  % Wypisanie wyniku

% Reguła uzupełniająca stan produktu
uzupelnij_stan(X, N) :-
    produkt(X, Ilosc, Sektor),
    NewIlosc is Ilosc + N,
    retract(produkt(X, Ilosc, Sektor)),  % Usuwa stary fakt
    assertz(produkt(X, NewIlosc, Sektor)),  % Dodaje nowy fakt z zaktualizowaną ilością
    write('produkt('), write(X), write(', '), write(NewIlosc), write(', '), write(Sektor), write(')'), nl.  % Wypisanie wyniku
```

```python
Przykładowe pytania i wyniki:
?- dostepny_produkt(tv).
Oczekiwany wynik: true
?- przenies_produkt(tv, sektor_b).
Oczekiwany wynik: produkt(tv, 10, sektor_b)
?- uzupelnij_stan(pralka, 3).
Oczekiwany wynik: produkt(pralka, 8, sektor_b)
```

```python
%Zad4
% Definiowanie miast
miasto(warszawa).
miasto(krakow).
miasto(poznan).
miasto(wroclaw).
miasto(gdansk).
miasto(lodz).
miasto(szczecin).
miasto(rzeszow).

% Definiowanie dróg między miastami oraz czasów przejazdu
droga(warszawa, krakow, 4).
droga(krakow, poznan, 3).
droga(warszawa, poznan, 7).
droga(krakow, wroclaw, 2).
droga(poznan, wroclaw, 3).
droga(wroclaw, gdansk, 6).
droga(warszawa, gdansk, 5).
droga(lodz, warszawa, 2).
droga(lodz, wroclaw, 3).
droga(szczecin, poznan, 4).
droga(szczecin, gdansk, 5).
droga(rzeszow, krakow, 3).
droga(rzeszow, warszawa, 6).

% Definiowanie drogi dwustronnej (jeśli istnieje droga X -> Y, to istnieje też Y -> X z tą samą długością)
dwustronna_droga(X, Y, C) :- droga(X, Y, C).    % X -> Y
dwustronna_droga(X, Y, C) :- droga(Y, X, C).    % Y -> X

% Predykat sprawdzający, czy istnieje droga między miastami X i Y (nie interesuje nas czas)
droga(X, Y) :- droga(X, Y, _), !.

% Predykat liczący czas przejazdu na najkrótszej trasie między miastami
czas_przejazdu(X, Y, T) :-
    najkrotsza_trasa(X, Y, _, T), !.

% Pomocniczy predykat do obliczenia najkrótszej trasy
min_trasa([Najlepsza | Reszta], Min) :- min_trasa(Reszta, Najlepsza, Min).

% Min_trasa, która porównuje czas podróży dla wszystkich tras i wybiera najkrótszą
min_trasa([], Min, Min).
min_trasa([(T, P) | Reszta], (NajlepszyT, NajlepszaP), Min) :-
    (T < NajlepszyT -> min_trasa(Reszta, (T, P), Min) ; min_trasa(Reszta, (NajlepszyT, NajlepszaP), Min)).

% DFS (przeszukiwanie w głąb) - sprawdza wszystkie możliwe trasy między miastami X i Y
dfs(Y, Y, Trasa, Czas, Trasa, Czas).  % Jeśli dotrzemy do celu, zwracamy trasę i czas

% Główna część DFS - przechodzi po drogach i rekurencyjnie szuka najkrótszej trasy
dfs(X, Y, Trasa, Czas, NajlepszaTrasa, NajkrotszyCzas) :-
    dwustronna_droga(X, Z, C),        % Znajdujemy drogę X -> Z
    \+ member(Z, Trasa),               % Sprawdzamy, czy Z już nie jest na trasie (unikamy cykli)
    NowyCzas is Czas + C,              % Dodajemy czas do bieżącego
    dfs(Z, Y, [Z | Trasa], NowyCzas, NajlepszaTrasa, NajkrotszyCzas).   % Rekurencja do Z

% Predykat do znalezienia najkrótszej trasy - wykonuje wszystkie możliwe przeszukiwania i wybiera najkrótszą
najkrotsza_trasa(X, Y, Trasa, Czas) :-
    findall((T, P), dfs(X, Y, [X], 0, P, T), Wyniki),   % Znajdujemy wszystkie możliwe trasy
    min_trasa(Wyniki, (Czas, Trasa)).   % Wybieramy najkrótszą trasę na podstawie czasu
```

```python
%Zad5
% Fakty o książkach (tytuł, autor, gatunek)
ksiazka('W pustyni i w puszczy', sienkiewicz, przygodowa).
ksiazka('Hobbit', tolkien, fantasy).
ksiazka('Pan Tadeusz', mickiewicz, poezja).

% Fakty o ocenach książek przez użytkowników (użytkownik, tytuł książki, ocena)
ocena(uzytkownik1, 'W pustyni i w puszczy', 5).
ocena(uzytkownik1, 'Hobbit', 4).
ocena(uzytkownik1, 'Pan Tadeusz', 3).

% Fakty o preferencjach gatunkowych użytkowników (użytkownik, preferowany gatunek)
preferencje(uzytkownik1, fantasy).

% Reguła do zbierania książek w określonym gatunku, które są wysoko oceniane przez użytkownika X
polecane_ksiazki(X, Gatunek, Ksiazki) :-
    preferencje(X, Gatunek),  % Użytkownik X ma preferencję na gatunek Gatunek
    findall(Tytul,
        (ksiazka(Tytul, _, Gatunek),         % Szukamy książek z tego gatunku
        ocena(X, Tytul, Ocena),              % Sprawdzamy ocenę książki przez użytkownika X
        Ocena >= 4),                        % Tylko książki ocenione na 4 lub więcej
        KsiazkiBezDuplikatow),              % Zbieramy tytuły książek
    list_to_set(KsiazkiBezDuplikatow, Ksiazki).  % Usuwamy duplikaty i zwracamy wynik
```
